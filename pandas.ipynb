### Stage 4

#### Task 1 - total runs per match

```python
#total runs per match
total_runs_per_match = deliveries_df.groupby('match_id')['total_runs'].sum()
for match_id, total_runs in total_runs_per_match.items():
    print(f"Match ID: {match_id}, Total Runs: {total_runs}")
```

#### Task 2 - Runs per Team per Match

```python
# Compute total runs scored by each team in every match.
print(deliveries_df.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index().sort_values(by='total_runs', ascending=False).head(10))
```

#### Task 3 - Top 10 batters

```python
top_batters = deliveries_df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
print(top_batters)
```

#### Task 4 - Strike Rate of Batters

```python
'''Calculate strike rate for each batter:

Runs scored
Balls faced
'''

runs_scored = deliveries_df.groupby('batter')['batsman_runs'].sum()
balls_faced = deliveries_df.groupby('batter')['ball'].count()

strike_rate = (runs_scored / balls_faced) * 100
print(strike_rate.sort_values(ascending=False).head(10))
```

#### Task 5 - Top 10 Bowlers by Economy

```python
# Calculate economy rate and rank bowlers.

bowler = deliveries_df.groupby('bowler')['total_runs'].sum()
# balls bowled should be greater than 0
balls_bowled = deliveries_df.groupby('bowler')['ball'].count() > 0
economy_rate = (bowler / (balls_bowled / 6))
print(economy_rate.sort_values().head(10))
```

#### Task 6

```python
'''Batters with High average runs per match'''

runs_per_match = deliveries_df.groupby(['batter', 'match_id'])['batsman_runs'].sum().reset_index()
average_runs_per_match = runs_per_match.groupby('batter')['batsman_runs'].mean().sort_values(ascending=False).head(10)
print(average_runs_per_match)
```

```python
# Minimum number of matches played (batters)
matches_played = deliveries_df.groupby('batter')['match_id'].nunique()
filtered_players = matches_played[matches_played < 10]
print(filtered_players)
```

#### Task 7

```python
# Find the highest runs scored by a batter in a single match.

runs_per_match = deliveries_df.groupby(['batter', 'match_id'])['batsman_runs'].sum().reset_index()
max_score = runs_per_match.loc[runs_per_match['batsman_runs'].idxmax()]

print(max_score)
```

#### Task 8

```python
''' Boundary Analysis
Total number of 4s and 6s
Top players by boundaries'''

total_fours = (deliveries_df['batsman_runs'] == 4).sum()
total_sixes = (deliveries_df['batsman_runs'] == 6).sum()

print("Total 4s:", total_fours)
print("Total 6s:", total_sixes)
```

#### Task 9

```python
total_runs = deliveries_df.groupby('batter')['batsman_runs'].sum()

boundary_runs = deliveries_df[deliveries_df['batsman_runs'].isin([4,6])] \
                    .groupby('batter')['batsman_runs'].sum()

boundary_percentage = (boundary_runs / total_runs) * 100
boundary_percentage = boundary_percentage.fillna(0).sort_values(ascending=False)

print(boundary_percentage.head(10))
```

```python
result = pd.DataFrame({
    'total_runs': total_runs,
    'boundary_runs': boundary_runs
})

result['boundary_percentage'] = (result['boundary_runs'] / result['total_runs']) * 100

result = result.fillna(0).sort_values(by='boundary_percentage', ascending=False)

print(result.head(10))
```

#### Task 16

```python
# Runs per batter per match
runs_per_match = deliveries_df.groupby(['match_id', 'batter'])['batsman_runs'].sum()

# Get top scorer per match
top_scorer = runs_per_match.groupby('match_id').idxmax()

# Convert to DataFrame
top_scorer = top_scorer.apply(lambda x: x[1]).reset_index(name='top_batter')

# Map player_of_match using dictionary
pom_map = matches_df.set_index('id')['player_of_match']

top_scorer['player_of_match'] = top_scorer['match_id'].map(pom_map)

# Compare
top_scorer['is_top_scorer'] = top_scorer['top_batter'] == top_scorer['player_of_match']

print(top_scorer['is_top_scorer'].value_counts())
```

#### Task 17

```python
venue_matches = matches_df['venue'].value_counts()
print(venue_matches)
```

```python
# total runs per match
match_runs = deliveries_df.groupby('match_id')['total_runs'].sum()

# map venue
venue_map = matches_df.set_index('id')['venue']

venue_runs = match_runs.to_frame('runs')
venue_runs['venue'] = venue_runs.index.map(venue_map)

avg_runs_venue = venue_runs.groupby('venue')['runs'].mean()

print(avg_runs_venue.sort_values(ascending=False))
```

#### Task 18

```python
city_map = matches_df.set_index('id')['city']

venue_runs['city'] = venue_runs.index.map(city_map)

avg_runs_city = venue_runs.groupby('city')['runs'].mean()

print(avg_runs_city.sort_values(ascending=False))
```

#### Task 19

```python
season_map = matches_df.set_index('id')['season']

venue_runs['season'] = venue_runs.index.map(season_map)

season_runs = venue_runs.groupby('season')['runs'].sum()

print(season_runs)
```

#### Task 20

```python
# total runs per team per match
match_runs = deliveries_df.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()

# predicted winner
predicted = match_runs.loc[
    match_runs.groupby('match_id')['total_runs'].idxmax()
]

# map actual winner
winner_map = matches_df.set_index('id')['winner']

predicted['actual_winner'] = predicted['match_id'].map(winner_map)

# compare
predicted['correct'] = predicted['batting_team'] == predicted['actual_winner']

accuracy = predicted['correct'].mean() * 100

print("Accuracy:", accuracy)
```
