```python
#%pip install numpy
import numpy as np
import csv
```

```python
deliveries_data= []
with open('data/deliveries.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        deliveries_data.append(row)
    
deliveries_array = np.array(deliveries_data)
```

```python
match_ids = deliveries_array[:, 0].astype(int)  # match_id as integers
batsman_runs = deliveries_array[:, 9].astype(int)  # batsman_runs as integers

# Using bincount for vectorized aggregation
unique_match_ids, counts = np.unique(match_ids, return_counts=True)
total_runs_per_match = np.bincount(match_ids, weights=batsman_runs)

# Filter to only matches that exist
result = list(zip(unique_match_ids, total_runs_per_match[unique_match_ids]))
for match_id, total_runs in result:
    print(f'Match ID: {match_id}, Total Runs: {total_runs}')
```

```python
# task 2 - top 5 batters

batsman_names = deliveries_array[:, 6]  # batsman names
batsman_runs = deliveries_array[:, 9].astype(int)  # batsman_runs
# Using bincount for vectorized aggregation
unique_batsmen, counts = np.unique(batsman_names, return_counts=True)
total_runs_per_batsman = np.bincount(np.searchsorted(unique_batsmen, batsman_names), weights=batsman_runs)

# Get top 5 batters
top_batsmen_indices = np.argsort(total_runs_per_batsman)[-5:]
top_batsmen = unique_batsmen[top_batsmen_indices]
top_runs = total_runs_per_batsman[top_batsmen_indices]
for batsman, runs in zip(top_batsmen, top_runs):
    print(f'Batsman: {batsman}, Total Runs: {runs}')

```

```python
batsman_indices = np.searchsorted(unique_batsmen, batsman_names)
balls_faced = np.bincount(batsman_indices)

# Strike rate calculation
strike_rate_per_batsman = (total_runs_per_batsman / balls_faced) * 100

# Get top 20 batters by strike rate
top_sr_indices = np.argsort(strike_rate_per_batsman)[-20:][::-1]
top_sr_batsmen = unique_batsmen[top_sr_indices]
top_strike_rates = strike_rate_per_batsman[top_sr_indices]

print("\nTop 20 Batters by Strike Rate:")
for batsman, sr in zip(top_sr_batsmen, top_strike_rates):
    print(f'Batsman: {batsman}, Strike Rate: {sr:.2f}')
```

```python

bowlers = deliveries_array[:, 7]  # bowler names
bowler_runs = deliveries_array[:, 11].astype(int)  # total runs conceded by bowler
bowler_indices = np.searchsorted(unique_batsmen, bowlers)
runs_conceded_per_bowler = np.bincount(bowler_indices, weights=bowler_runs)
balls_bowled_per_bowler = np.bincount(bowler_indices)
economy_rate_per_bowler = runs_conceded_per_bowler / (balls_bowled_per_bowler / 6)

# Get top 20 bowlers by economy rate
top_economy_indices = np.argsort(economy_rate_per_bowler)[:20]
top_economy_bowlers = unique_batsmen[top_economy_indices]
top_economy_rates = economy_rate_per_bowler[top_economy_indices]
print("\nTop 20 Bowlers by Economy Rate:")
for bowler, economy in zip(top_economy_bowlers, top_economy_rates):
    print(f'Bowler: {bowler}, Economy Rate: {economy:.2f}')
```
