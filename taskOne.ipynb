```python
#%pip install numpy
import numpy as np
import csv
```

```python
deliveries_data= []
with open('data/deliveries.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        deliveries_data.append(row)
    
deliveries_array = np.array(deliveries_data)
```

Task 1

```python
match_ids = deliveries_array[:, 0].astype(int)  # match_id as integers
batsman_runs = deliveries_array[:, 9].astype(int)  # batsman_runs as integers

# Using bincount for vectorized aggregation
unique_match_ids, counts = np.unique(match_ids, return_counts=True)
total_runs_per_match = np.bincount(match_ids, weights=batsman_runs)

# Filter to only matches that exist
result = list(zip(unique_match_ids, total_runs_per_match[unique_match_ids]))
for match_id, total_runs in result:
    print(f'Match ID: {match_id}, Total Runs: {total_runs}')
```

Task 2

```python
# task 2 - top 5 batters

batsman_names = deliveries_array[:, 6]  # batsman names
batsman_runs = deliveries_array[:, 9].astype(int)  # batsman_runs
# Using bincount for vectorized aggregation
unique_batsmen, counts = np.unique(batsman_names, return_counts=True)
total_runs_per_batsman = np.bincount(np.searchsorted(unique_batsmen, batsman_names), weights=batsman_runs)

# Get top 5 batters
top_batsmen_indices = np.argsort(total_runs_per_batsman)[-5:]
top_batsmen = unique_batsmen[top_batsmen_indices]
top_runs = total_runs_per_batsman[top_batsmen_indices]
for batsman, runs in zip(top_batsmen, top_runs):
    print(f'Batsman: {batsman}, Total Runs: {runs}')

```

Task 3

```python
batsman_indices = np.searchsorted(unique_batsmen, batsman_names)
balls_faced = np.bincount(batsman_indices)

# Strike rate calculation
strike_rate_per_batsman = (total_runs_per_batsman / balls_faced) * 100

# Get top 20 batters by strike rate
top_sr_indices = np.argsort(strike_rate_per_batsman)[-20:][::-1]
top_sr_batsmen = unique_batsmen[top_sr_indices]
top_strike_rates = strike_rate_per_batsman[top_sr_indices]

print("\nTop 20 Batters by Strike Rate:")
for batsman, sr in zip(top_sr_batsmen, top_strike_rates):
    print(f'Batsman: {batsman}, Strike Rate: {sr:.2f}')
```

Task 4

```python

bowlers = deliveries_array[:, 7]  # bowler names
bowler_runs = deliveries_array[:, 11].astype(int)  # total runs conceded by bowler
bowler_indices = np.searchsorted(unique_batsmen, bowlers)
runs_conceded_per_bowler = np.bincount(bowler_indices, weights=bowler_runs)
balls_bowled_per_bowler = np.bincount(bowler_indices)
economy_rate_per_bowler = runs_conceded_per_bowler / (balls_bowled_per_bowler / 6)

# Get top 20 bowlers by economy rate
top_economy_indices = np.argsort(economy_rate_per_bowler)[:20]
top_economy_bowlers = unique_batsmen[top_economy_indices]
top_economy_rates = economy_rate_per_bowler[top_economy_indices]
print("\nTop 20 Bowlers by Economy Rate:")
for bowler, economy in zip(top_economy_bowlers, top_economy_rates):
    print(f'Bowler: {bowler}, Economy Rate: {economy:.2f}')
```

Task 5

```python
overs = deliveries_array[:, 4].astype(int)
batsman_runs = deliveries_array[:, 9].astype(int)
match_ids = deliveries_array[:, 0].astype(int)
innings = deliveries_array[:, 1].astype(int)      # column 1 = inning

avg_runs_per_over = np.zeros(20, dtype=float)

for over in range(20):
    over_mask = (overs == over)
    
    # Group by (match_id, inning) pairs to calculate total runs for each over in each match and inning
    match_inning_pairs = np.unique(
        np.column_stack((match_ids[over_mask], innings[over_mask])), axis=0
    )
    
    runs_per_innings = np.array([
        batsman_runs[over_mask & (match_ids == m) & (innings == ing)].sum()
        for m, ing in match_inning_pairs
    ])
    
    avg_runs_per_over[over] = runs_per_innings.mean() if len(runs_per_innings) > 0 else 0.0

print("Average Runs per Over (across all matches):")
for i, avg in enumerate(avg_runs_per_over, start=1):
    print(f"Over {i}: {avg:.2f} runs")
```

Task 6

```python
# Extract relevant columns
batsman_runs = deliveries_array[:, 9].astype(int)  
batting_team = deliveries_array[:, 2]               

total_fours = np.sum(batsman_runs == 4)
total_sixes = np.sum(batsman_runs == 6)

print(f"Total number of 4s: {total_fours}")
print(f"Total number of 6s: {total_sixes}")

unique_teams = np.unique(batting_team)

team_boundaries = {}
for team in unique_teams:
    team_mask = (batting_team == team)
    team_runs  = batsman_runs[team_mask]
    boundaries = np.sum(team_runs == 4) + np.sum(team_runs == 6)
    team_boundaries[team] = boundaries

# Find team with max boundaries
best_team = max(team_boundaries, key=lambda t: team_boundaries[t])

print("\nBoundaries per team:")
for team, count in sorted(team_boundaries.items(), key=lambda x: -x[1]):
    print(f"  {team}: {count}")

print(f"\nTeam with most boundaries: {best_team} ({team_boundaries[best_team]} boundaries)")
```

Task 7

```python
# Extract overs (column 4), total runs (column 11), and batting team (column 2)
overs = deliveries_array[:, 4].astype(int)
total_runs = deliveries_array[:, 11].astype(int)
batting_team = deliveries_array[:, 2]

# Mask for death overs (16–20)
death_overs_mask = (overs >= 16) & (overs <= 20)

# Total runs scored in death overs
total_death_runs = np.sum(total_runs[death_overs_mask])
print(f"\nTotal runs scored in death overs (16–20): {total_death_runs}")

# Runs per team in death overs
death_batting_team = batting_team[death_overs_mask]
death_runs = total_runs[death_overs_mask]

unique_teams = np.unique(death_batting_team)
death_runs_per_team = {}

for team in unique_teams:
    team_mask = (death_batting_team == team)
    team_runs = np.sum(death_runs[team_mask])
    death_runs_per_team[team] = team_runs

# Team with highest runs in death overs
best_death_team = max(death_runs_per_team, key=lambda t: death_runs_per_team[t])

print("\nDeath overs runs per team:")
for team, runs in sorted(death_runs_per_team.items(), key=lambda x: -x[1]):
    print(f"  {team}: {runs} runs")

print(f"\nTeam with highest death overs runs: {best_death_team}")
```

Task 8

```python
match_ids = deliveries_array[:, 0].astype(int)
total_runs = deliveries_array[:, 11].astype(int)

# Get unique match ids and sum runs per match
unique_match_ids = np.unique(match_ids)

runs_per_match = np.array([
    total_runs[match_ids == m].sum()
    for m in unique_match_ids
])

# Find match with highest total runs
max_index = np.argmax(runs_per_match)
best_match_id = unique_match_ids[max_index]
best_match_runs = runs_per_match[max_index]

print(f"Match with highest total runs: Match ID {best_match_id} with {best_match_runs} runs")
```

Task 9

```python
'''Task 9: Runs per Team per Match
Objective
Calculate total runs scored by each team in each match.
Hint
You may need to combine:
match_id
batting_team
'''

match_ids = deliveries_array[:, 0].astype(int)
batting_team = deliveries_array[:, 2]
total_runs = deliveries_array[:, 11].astype(int)
# Get unique match ids and teams
unique_match_ids = np.unique(match_ids)
unique_teams = np.unique(batting_team)
runs_per_team_per_match = {}

for match_id in unique_match_ids:
    match_mask = (match_ids == match_id)
    teams_in_match = np.unique(batting_team[match_mask])
    runs_per_team = {}
    for team in teams_in_match:
        team_mask = match_mask & (batting_team == team)
        team_runs = np.sum(total_runs[team_mask])
        runs_per_team[team] = team_runs
    runs_per_team_per_match[match_id] = runs_per_team

# Print runs per team per match
for match_id, team_runs in runs_per_team_per_match.items():
    print(f"Match ID {match_id}:")
    for team, runs in team_runs.items():
        print(f"  {team}: {runs} runs")
```
